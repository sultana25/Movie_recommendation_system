```python
import numpy as np
import pandas as pd
import ast
import nltk
```

```python
movies = pd.read_csv('dataset/tmdb_5000_movies.csv')
credits = pd.read_csv('dataset/tmdb_5000_credits.csv')
```

```python
movies = movies.merge(credits, on = "title")
#movies['original_language'].value_counts()
#movies.head(1)
```

```python
#movie_id
#genres
#keywords
#title
#cast
#crew
#movies.info()
movies = movies[["movie_id","title","keywords","genres","overview","cast","crew"]]
#movies.head(1)
movies.dropna(inplace = True)
#movies.isnull().sum()
#movies.iloc[0].genres
```

```python
#helper function to convert the format genres into list
def convert(obj):
    L = []
    for i in ast.literal_eval(obj):
        L.append(i['name'])
    return L

```

```python
#import ast
#ast.literal_eval('[{"id": 28, "name": "Action"}, {"id": 12, "name": "Adventure"}, {"id": 14, "name": "Fantasy"}, {"id": 878, "name": "Science Fiction"}]')
```

```python
movies['genres'] = movies['genres'].apply(convert)
```

```python
#movies.head()
movies['keywords'] = movies['keywords'].apply(convert)
```

```python
movies.head(1)
```

```python
#to convert string to list
def convert2(obj):
    L = []
    counter = 0
    for i in ast.literal_eval(obj):
        if counter!= 3:
            L.append(i['name'])
            counter+=1
        else:
            break
    return L
```

```python
movies['cast'] = movies['cast'].apply(convert2)
```

```python
#to fetch the director name
def fetch_director(obj):
    L=[]
    for i in ast.literal_eval(obj):
        if i['job']=='Director':
            L.append(i['name'])
            break
    return L       
```

```python
movies['crew'] = movies['crew'].apply(fetch_director)
```

```python
movies['overview'] = movies['overview'].apply(lambda x:x.split())
```

```python
#to
movies['genres'] = movies['genres'].apply(lambda x:[i.replace(" ","") for i in x])
movies['keywords'] = movies['keywords'].apply(lambda x:[i.replace(" ","") for i in x])
movies['cast'] = movies['cast'].apply(lambda x:[i.replace(" ","") for i in x])
movies['crew'] = movies['crew'].apply(lambda x:[i.replace(" ","") for i in x])
```

```python
movies['tags'] = movies['overview']+movies['genres']+movies['keywords']+movies['cast']+movies['crew']
```

```python
movies_df = movies[['movie_id','title','tags']] 
```

```python
movies_df['tags'] = movies_df['tags'].apply(lambda x:" ".join(x))
```

```python
movies_df['tags'] = movies_df['tags'].apply(lambda x:x.lower())
```

```python
movies_df.head()
```

```python
from sklearn.feature_extraction.text import CountVectorizer
cv = CountVectorizer(stop_words='english',max_features=5000)
```

```python
vectors = cv.fit_transform(movies_df['tags']).toarray()
```

```python
cv.get_feature_names()
```

```python
from nltk.stem.porter import PorterStemmer
# to reduce words to their root or base form.
ps = PorterStemmer()
```

```python
#helper function 
def stem(text):
    y = []
    for i in text.split():
        y.append(ps.stem(i))
        
    return " ".join(y)
        
```

```python
movies_df['tags'] = movies_df['tags'].apply(stem)
```

```python
#movies_df['tags'][0]
```

```python
from sklearn.metrics.pairwise import cosine_similarity
similarity = cosine_similarity(vectors)
```

```python
#similarity[2]
```

```python
#movies_df[movies_df['title']== 'Avatar'].index[0]
#sorted(list(enumerate(similarity[0])), reverse = True, key=lambda x:x[1])[1:6]
```

```python
def recommend(movie):
    movie_index = movies_df[movies_df['title']==movie].index[0]
    distance = similarity[movie_index]
    movie_list = sorted(list(enumerate(distance)), reverse = True, key=lambda x:x[1])[1:6]
    for i in movie_list:
        print(movies_df.iloc[i[0]].title)
    
```

```python
recommend('Batman')
```
